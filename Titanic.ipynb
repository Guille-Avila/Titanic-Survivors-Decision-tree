```python
import pandas as pd
data = pd.read_csv('Titanic_data/train.csv')
test = pd.read_csv('Titanic_data/test.csv')
data.head()
```

```python
inputs = data.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], axis='columns')
target = data['Survived']
```

```python
target.head()
```

## Preparing Data

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
le_Sex = LabelEncoder()
le_Embarked = LabelEncoder()
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
inputs['Embarked_n'] = le_Embarked.fit_transform(inputs['Embarked'])
inputs.head()
```

```python
inputs_n = inputs.drop(['Sex', 'Embarked'], axis='columns')
inputs_n.head(20)
```

```python
'''import numpy as np

inputs_n['Age'] = inputs_n['Age'].replace('NaN', np.nan)

inputs_n['Age'] = inputs_n['Age'].fillna(int(inputs_n['Age'].mean()))
inputs_n.head(20)'''

inputs_n.Age = inputs_n.Age.fillna(inputs_n.Age.mean())

data.Fare = data.Fare.fillna(data.Fare.mean())
test.Fare = test.Fare.fillna(test.Fare.mean())
data.Age = data.Age.fillna(data.Age.mean())
test.Age = test.Age.fillna(test.Age.mean())
inputs_n.head()
```

## Model

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n, target)
```

```python
inputs_n.info()
```

```python
le_Sex_test = LabelEncoder()
le_Embarked_test = LabelEncoder()
test['Sex_n'] = le_Sex_test.fit_transform(test['Sex'])
test['Embarked_n'] = le_Embarked_test.fit_transform(test['Embarked'])
test.head()
```

```python
features = ["Pclass", 'Age', "SibSp", "Parch", 'Fare', "Sex_n", 'Embarked_n']
X_test = pd.get_dummies(test[features])
X_test.head(10)
```

```python
predictions = model.predict(X_test)
```
